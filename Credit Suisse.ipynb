```python
#Importing libraries: Pandas, seaborn, numpy and matplotlib
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

#Setting plt to use the ggplot style
plt.style.use('ggplot')

#Importing the 'figure' function
from matplotlib.pyplot import figure

#Inline is a jupyter command for matplotlib that sets Jupyter to return the plots in the output, rather than opening a new window
%matplotlib inline

matplotlib.rcParams['figure.figsize']=(15,8)

#Importing the CSV file
#Note: To get the file path, go to the file location, go to properties, copy the path from there and then add the file name
#Preferably by clicking F2 and copying the name
#Note2: Don't forget to add the '.csv' at the end of the path
#Note3: Don't forget to add an 'r' at the beginning of the path, after the parenthesis

#Credit Suisse
df = pd.read_csv(r'C:\Users\Juan\Desktop\DS\CS.csv')
```

```python
df.head()
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
df.dtypes
```

```python
df.dropna(inplace=True) # Remove missing values
df.drop_duplicates(inplace=True) # Remove duplicates
```

```python
# Create a candlestick chart
fig, ax = plt.subplots()
ax.set_title('Candlestick Chart for Credit Suisse Stock Prices 2017-2023')
ax.plot(df.index, df['High'], label='High')
ax.plot(df.index, df['Low'], label='Low')
ax.vlines(df.index, df['Open'], df['Close'], color='black', alpha=0.4, linewidth=2)
ax.legend()
```
